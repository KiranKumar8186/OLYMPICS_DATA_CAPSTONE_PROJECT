### Q1. Write PySpark code to read olympix_data.csv file 

```python
# CONVERTING PANDAS DATFRAME TO PYSPARK DATAFRAME
# PRINTING FIRST 20 OBSERVATIONS IN PANDAS FORMAT TO GET BETTER VISUALIZATION
# USING CLEANED DATA THAT WE HAVE DONE IN PYTHON AND DS

olympic_spark = spark.createDataFrame(Olympic_pandas)
olympic_spark.limit(20).toPandas()
```

```python
# PRINTING THE SCHEMA TO GET THE ALL DATA TYPES OF THE DATA SET

olympic_spark.printSchema()
```

```python
# COUNTING THE TOTAL NUMBER OF ROWS PRESENT IN THE DATA
row = olympic_spark.count()

# AND ALSO FINDING THE TOTAL NUMBER OF COLUMNS IN THE DATA
column = len(olympic_spark.columns)

# PRINTING THE TOTAL NUMBER OF ROWS AND COLUMNS THAT ARE IN THE DATASET 
print(f'Dimension of the Dataframe is: {(row,column)}')
```

###### Q2. Write PySpark code to print the Olympic Sports/games in the dataset.

```python
# USING SELECT FUNCTION TO SELECT THE 'SPORTS' COLUMN 
# AND PRINTING ALL OLYMPIC SPORTS NAMES IN PANDAS FORMAT

olympic_spark.select("sports").distinct().show()
```

###### Q3. Write PySpark code to plot the total number of medals in  each Olympic Sport/game

```python
# USING GROUP BY FUNCTION TO SPORTS COLUMN TO GET UNIQUE NAME (THERE ARE SPORTS NAME HAVE BEEN REPEATED)
# AND CALCULATING SUM OF TOTAL_MEDALS
# TO GET THE TOTAL NUMBER OF MEDALS IN EACH OLYMPIC SPORT and PRINTING FIRST 10 RECORDS IN PANDAS FORMAT

olympic_total_medal = olympic_spark.groupBy('sports').agg(sum('total_medal').alias('total_medals'))
olympic_total_medal.limit(10).toPandas()
```

###### Q4. Sort the result based on the total number of medals.


```python
#  HERE SORTING THE TOTAL MEDALS IN EACH OLYMPIC SPORT IN DECREASING ORDER FROM ABOVE ANSWER
# AND PRINTING FIRST 10 RECORDS IN PANDAS FORMAT

olympic_total_medal.sort(desc('total_medals')).limit(10).toPandas()
```

###### Q5. Find the total number of medals won by each country in swimming.


```python
# ASSIGNING THE SPORT NAME AS 'Swimming'
# AND USING GROUP BY FUNCTION TO COUNTRY AND SPORTS COLUMN 
# CALCULATING THE SUM OF TOTAL_MEDALS TO 
# GETTING THE TOTAL NUMBER OF MEDALS WON BY EACH COUNTRY IN SWIMMING
# AND PRINTING FIRST 20 RECORDS IN PANDAS FORMAT

olympic_spark[olympic_spark['sports'] == 'Swimming'].groupby('country', 'sports').agg(sum('total_medal').alias('total_medals')).limit(20).toPandas()
```

###### Q6. Find the total number of medals won by each country in Skeleton.


```python
# ASSIGNING THE SPORT NAME AS 'Skeleton'
# AND USING GROUP BY FUNCTION TO COUNTRY AND SPORTS COLUMN 
# CALCULATING THE SUM OF TOTAL_MEDALS TO 
# GETTING THE TOTAL NUMBER OF MEDALS WON BY EACH COUNTRY IN SWIMMING
# AND PRINTING FIRST 10 RECORDS IN PANDAS FORMAT

olympic_spark[olympic_spark['sports'] == 'Skeleton'].groupby('country', 'sports').sum('total_medal').limit(10).toPandas()
```

###### Q7. Find the number of medals that the US won yearly.


```python
# ASSIGNING THE 'COUNTRY' NAME AS 'United States'
# AND USING GROUP BY FUNCTION TO COUNTRY AND YEAR COLUMN
# CALCULATING THE SUM OF TOTAL_MEDALS TO 
# GETTING THE TOTAL NUMBER OF MEDALS WON BY EACH COUNTRY IN SWIMMING
# AND PRINTING IN PANDAS FORMAT

olympic_spark[olympic_spark['country'] == 'United States'].groupby('country', 'year').agg(sum('total_medal').alias('total_medals')).limit(10).toPandas()
```

###### Q8. Find the total number of medals won by each country.


```python
# GROUP BY FUNCTION USING TO COUNTRY COLUMN 
# AND CALCULATING THE SUM OF TOTAL_MEDALS TO
# FIND THE TOTAL NUMBER OF MEDALS WON BY EACH COUNTRY
# AND PRINTING FIRST 10 VALUES IN PANDAS FORMAT 

olympic_spark.groupby('country').sum('total_medal').limit(10).toPandas()
```

###### Q9. Who was the oldest athlete in the olympics? 


```python
# USING GROUP BY FUNCTION TO NAME COLUMN AND 
# FINDING THE MAX AGE OF ATHLETES WHEN THEY HAVE PLAYED WON LAST AND SAVING IN A DATA

oldest_athlete = olympic_spark.groupBy('name').max('age')
```

```python
# FROM ABOVE SAVED DATA SORTING THE VALUES OF MAX AGE AND TAKING THAT TOP 1 ROW
# WHICH WILL BE THE OLDEST ATHLETE 

oldest_athlete.sort(desc('max(age)')).show(1)
```

###### Q10. Which country was he/she from?


```python
# UNSING SELECT FUNCTION TO NAME AND COUNTRY COLUMNS TO GET UNIQUE NAMES 
# PRINTING THAT WHICH COUNTRY  WAS HE OR SHE BELONGS
# AND PRINTING FIRST 2O RECORDS IN PANDAS FORMAT 

olympic_spark.select('name', 'country').distinct().limit(20).toPandas()
```
